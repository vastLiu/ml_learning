```python
import numpy as np
import matplotlib as mpl

import matplotlib.pyplot as plt

## 原函数
def f(x):
    return x**2+2*x

def h(x):
    return 2*x+2

X = []
Y = []

x = 5.5
step = 0.9

f_change = f(x)
f_current = f_change
X.append(x)
Y.append(f_current)
iter_num = 0
while f_change > 1e-10 and iter_num < 100:
    x = x - step*h(x)
    f_change = np.abs(f(x)-f_current)
    f_current = f(x)
    X.append(x)
    Y.append(f_current)
    iter_num += 1
print(u'最终结果:',(x,f_current))
print(u'迭代次数',iter_num)


# print([x, y for x, y in zip(X,Y)])
[print([x,y]) for x, y in zip(X,Y)]

%matplotlib tk

#画图
fig = plt.figure()
X0 = np.arange(-6,6,0.05)
Y0 = X0**2+2*X0

plt.plot(X0,Y0,'-')
plt.plot(X,Y,'r-.o')
plt.show()

```
