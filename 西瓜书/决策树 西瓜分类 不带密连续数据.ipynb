```python
datas = datas.replace(['是','否'],[1,0])
datas = datas.drop(columns=['密度','含糖率'])
datas
```

```python

def Ent(Data):
    # 计算Data的的信息熵，统计在Data数据中的正负例的数目
    res = 0.0
    for i in Data['好瓜'].value_counts():
        res += -(i/len(Data))*np.log2(i/len(Data))
    return res


def split_data_by_attr(Data, attr):
    data_values = Data[attr].value_counts()
    values = data_values.index
    value_list = []
    
    for value in values:
        value_list.append([value,[]])
    value_dict = dict(value_list)
    for index in range(len(Data)):
        value_dict[Data.iloc[index][attr]].append(index) 
    print(value_dict)
    return value_dict

        
        
        
def Gain(Data, attr):
    #先对attr进行Data的划分
    Ent_remain = Ent(Data)
    data_dict = split_data_by_attr(Data, attr)
    
    sum_num = sum([len(v) for v in data_dict.values()])
    for k,v in data_dict.items():
        Ent_remain -= Ent(Data.iloc[v])*len(v)/sum_num
#     for i in Data[attr].value_counts:
#         dt.add()     
    return Ent_remain
    
```

```python
## 
## 开始决策树的构造
## 每一部构造，打印出叶子节点以及该叶子节点对应的属性
# [1,[1,2],[[1,2],[3,4]]]

def construct_decision_tree(Data):
    print(Data)
    #1.找到Data的所有属性：
    if len(Data)<2:
        return 
    names = Data.columns
    if len(names) == 1:
        return 
    #2.计算所有属性的增益
    gains = {name:Gain(Data, name) if name not in ['密度','含糖率'] else 1 for name in names[:-1]}
    print(gains)
    #3.取出最大的增益对应的属性
    max_gain = 0.0
    max_key = ''
    for k,v in gains.items():
        if v > max_gain:
            max_gain = v
            max_key = k
    print(max_key,max_gain)
    if max_gain < 1e-5:
        return 
    #4.对该属性进行数据的分割
    sub_trees = split_data_by_attr(Data, max_key)
    print('节点->'+max_key)
    for v in sub_trees.values():
        construct_decision_tree(Data.iloc[v].drop(columns = [max_key]))
```
